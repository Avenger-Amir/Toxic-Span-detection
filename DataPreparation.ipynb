```python
def getY_word(XT,YT):
    Y_word=[]
    for i in range(0,len(YT)):
        XT[i]=XT[i].replace("-"," ")
        tmpY=YT[i][1:-1]
        if len(tmpY)>0:
            tmpY=list(map(int,tmpY.split(", ")))
            s=""
            l=[]
            prev=0
            for idx in tmpY:
                if (XT[i][int(idx)].lower()>='a' and XT[i][int(idx)].lower()<='z') or (XT[i][idx]=='\'') or (XT[i][idx]=='*'):
                    if prev!=idx-1:
                        if len(s)>0:
                            l.append(s)
                            s=""
                    s+=XT[i][int(idx)]
                else:
                    if len(s)>0:
                        l.append(s)
                    s=""
                prev=idx

            if len(s)>0:
                l.append(s)
            #print(f'{l} {i}')
            Y_word.append(l)
        else:
            l=[]
            Y_word.append(l)
    return Y_word
```

```python
def removePunctuation2(Y_word):
    YWordWithoutPunctuation=[]
    for i in range(0,len(Y_word)):
        l=[]
        for j in range(0,len(Y_word[i])):
            s=""
            for k in range(0,len(Y_word[i][j])):
                if (Y_word[i][j][k].lower()>='a' and Y_word[i][j][k].lower()<='z') or(Y_word[i][j][k]==' '):
                    s+=Y_word[i][j][k].lower()
            l.append(s)
        YWordWithoutPunctuation.append(l)
    return YWordWithoutPunctuation
```

```python
def removePunctuation1(XT):
    XTWithoutPunctuation=[0]*(len(XT))
    for i in range(0,len(XT)):
        s=""
        for j in range(0,len(XT[i])):
            if (XT[i][j].lower()>='a' and XT[i][j].lower()<='z') or(XT[i][j]==' '):
                s+=XT[i][j].lower()
        XTWithoutPunctuation[i]=s
    return XTWithoutPunctuation
```

```python
def stopWordRemoval1(XTWithoutPunctuation,stopWords):
    XTWithoutStopWord=[0]*len(XTWithoutPunctuation)
    for i in range(0,len(XTWithoutPunctuation)):
        s=XTWithoutPunctuation[i]
        s=s.split()
        l=[]
        for word in s:
            if word not in stopWords:
                l.append(word)
        s=" ".join(l)
        XTWithoutStopWord[i]=s
    return XTWithoutStopWord
```

```python
def stopWordRemoval2(YWordWithoutPunctuation,stopWords):
    YWithoutStopWord=[]
    for i in range(0,len(YWordWithoutPunctuation)):
        l=[]
        for word in YWordWithoutPunctuation[i]:
            if word not in stopWords:
                l.append(word)
        YWithoutStopWord.append(l)
    return YWithoutStopWord
```

```python
def snowBallStemming1(XTWithoutStopWord,stemmer):
    XTStemmed=[]
    for i in range(0,len(XTWithoutStopWord)):
        s=XTWithoutStopWord[i]
        s=s.split()
        sent=""
        for word in s:
            stem = stemmer.stem(word)
            sent+=stem
            sent+=" "
        XTStemmed.append(sent)
    return XTStemmed
```

```python
def snowBallStemming2(YWithoutStopWord,stemmer):
    YTStemmed=[]
    for i in range(0,len(YWithoutStopWord)):
        l=[]
        for word in YWithoutStopWord[i]:
            stem=stemmer.stem(word)
            l.append(stem)
        YTStemmed.append(l)
    return YTStemmed
```

```python
def doIndexing(XTStemmed,YTStemmed,maxLen):
    YIndex=np.zeros((len(XTStemmed),maxLen))
    for i in range(0,len(XTStemmed)):
        previdx=-1
        for yword in YTStemmed[i]:
            s=XTStemmed[i];
            s=s.split()
            for j in range(0,len(s)):
                if s[j]==yword and previdx<j:
                    YIndex[i][j]=1
                    previdx=j
                    break
    return YIndex
```

```python
def XEmbedding(XTStemmed,maxLen,embeddings):
    tmpArray=np.zeros((50))
    embedding_matrix_output = np.zeros((len(XTStemmed),maxLen,50))
    print(embedding_matrix_output.shape)
    for ix in range(len(XTStemmed)):
        s=XTStemmed[ix]
        s=s.split()
        for jx in range(len(s)):
            if s[jx] not in embeddings:
                embedding_matrix_output[ix][jx]=tmpArray
                continue
            embedding_matrix_output[ix][jx] = embeddings[s[jx]]
            
    return embedding_matrix_output
```

```python
def get_idx(s,idx,id):
    cnt=0
    idx=int(idx)
    for j in range(0,idx+1):
        if s[j]==' ':
            cnt+=1
            
    tmpcnt=0
    for j in range(0,len(s)):
        if s[j]==' ':
            tmpcnt+=1
    
    if id==20:
        for j in range(0,len(s)):
            print(f'{j} {s[j]}')
    if(len(s.split())!=tmpcnt):
        if(tmpcnt<10):
            print(s.split())
            print(s)
            print(f'{id} {len(s.split())} {tmpcnt}')
    return cnt
```

```python
def get_YT(XT,YT):
    emb_YT=np.zeros((YT.shape[0],256))
    for i in range(0,YT.shape[0]):
        YT[i]=YT[i][1:-1]
        YT[i]=YT[i].split(', ')
        #print(YT[i],len(YT[i]))
        if len(YT[i])<=1:
            continue
        for j in range(0,len(YT[i])):
            idx=get_idx(XT[i],YT[i][j],i)
            emb_YT[i][idx]=1
    return emb_YT
```

```python
def lancaster_stemmer(word):
    lancaster=LancasterStemmer()
    new_word=""
    for ch in word:
        if ch>='a' and ch<='z':
            new_word+=ch
    return lancaster(new_word)
```

```python
tmpArray=np.zeros((50))
def getOutputEmbeddings(X):
    embedding_matrix_output = np.zeros((X.shape[0],256,50))
    for ix in range(X.shape[0]):
        X[ix] = X[ix].split()
#         print(X[ix])
        for jx in range(len(X[ix])):
            processed_word=lancaster_stemmer(X[ix][jx].lower)
            if(embeddings.get(X[ix][jx].lower)==None):
                embedding_matrix_output[ix][jx]=tmpArray
                continue
            embedding_matrix_output[ix][jx] = embeddings[X[ix][jx].lower()]
            
    return embedding_matrix_output
```
